```python
# FUNCTION TO PLOT SINGLE INSTANTANEOUS MAPS -  to be externalised

def single_plot(alti_20,min_sl,max_sl,counter,fig):

    lon = alti_20.lon.values
    lat = alti_20.lat.values
    ssh = alti_20.sla_predicted.values
    min_lat=53.0
    max_lat=66.0
    min_lon=9.0
    max_lon=31.0  
    
    def plot_cities(ax):
        # plot city dots with annotation, finalize plot
        # lat/lon coordinates of five cities in Bangladesh
        lats = [55.4167, 56.1053, 55.5575]
        lons = [12.8294, 15.5894, 14.3578]
        cities = ['Sk', 'Ku', 'Si']
        for lon, lat, city in zip(lons, lats, cities):
            ax.plot(lon, lat, 'ro', zorder=5, transform=ccrs.PlateCarree())
            ax.text(lon + 0.01, lat + 0.01, city, fontsize='large',
                    transform=ccrs.PlateCarree()) 
    
    
    """Displaying the unstructured grid in a scatterplot:"""
    # 

    #fig = plt.plot()
    #plt.rcParams.update({'font.size': 15})
    #plt.plot
    #plt.rcParams["figure.figsize"] = (50,10) #Increase figure size
    ax=fig.add_subplot(1,3,counter,projection=ccrs.PlateCarree())
    
    ax.set_xticks(np.arange(min_lon,max_lon,2), crs=ccrs.PlateCarree())
    ax.set_yticks(np.arange(min_lat,max_lat,1), crs=ccrs.PlateCarree())    
    
    #ax = plt.axes(projection=ccrs.Miller()) 
    img=plt.scatter(lon, lat, c=ssh, s=20, cmap='cool', alpha=1) 
    ax.coastlines(resolution='10m', color='black', linewidth=1) 
    plt.xlim(min_lon,max_lon)

    lon_formatter = cart.mpl.ticker.LongitudeFormatter(number_format='.1f', 
                                      degree_symbol='', 
                                      dateline_direction_label=True)
    lat_formatter = cart.mpl.ticker.LatitudeFormatter(number_format='.1f', 
                                     degree_symbol='')
    ax.xaxis.set_major_formatter(lon_formatter)
    ax.yaxis.set_major_formatter(lat_formatter) 
    plt.colorbar(img,label=r'SSH (m)',orientation = 'horizontal',pad = 0.1)
    plt.clim(min_sl, max_sl)
    
    plot_cities(ax)
    
    #plt.show() #a window will show up, allowing the user to see and download the plot
    
    return 
```

## Compute correlation between CMEMS Altimetry and Copernicus reanalysis

```python
def corr_with_model(groups, groups_cmems, corr_copernicus_cmems_plot, var_cmems_plot) :
    counter = 0
    for group_key, group_value in groups:
        counter = counter+1
        group = groups.get_group(group_key)
        try:
            group_cmems = groups_cmems.get_group(group_key)
        except:
            continue
        #plt.plot(group.sla_predicted)
        #plt.plot(alti_cmems.sla_predicted[group.times])    
        #print(group)
        #print("")
        lat_corr_plot = group_key[0]
        lon_corr_plot = group_key[1]
        try:
            corr_copernicus_cmems = np.corrcoef(group.sla_predicted[0:182], group_cmems[0:182].sla_predicted)[0,1]    
            corr_copernicus_cmems_plot[counter-1,:] = np.array([lat_corr_plot,lon_corr_plot,corr_copernicus_cmems])
        
            var_cmems = np.var(group_cmems[0:182].sla_predicted)
            var_cmems_plot[counter-1,:] = np.array([lat_corr_plot,lon_corr_plot,var_cmems])
        except:
           continue    
        #print(corr_copernicus_cmems)
        #break
    
    return corr_copernicus_cmems_plot, var_cmems_plot




groups = alti_Copernicus.groupby(['lat', 'lon'])
groups_cmems = alti_cmems.groupby(['lat', 'lon'])
groups_ML = alti_ML.groupby(['lat', 'lon'])

#Correlations
corr_copernicus_cmems_plot = np.zeros([np.shape(groups)[0], 3])
corr_copernicus_ML_plot = np.zeros([np.shape(groups)[0], 3])

#Variances
var_cmems_plot = np.zeros([np.shape(groups)[0], 3])
var_ML_plot = np.zeros([np.shape(groups)[0], 3])


counter = 0

corr_copernicus_cmems_plot, var_cmems_plot \
= corr_with_model(groups, groups_cmems, corr_copernicus_cmems_plot, var_cmems_plot)

corr_copernicus_ML_plot, var_ML_plot \
= corr_with_model(groups, groups_ML, corr_copernicus_ML_plot, var_ML_plot)

    
```
